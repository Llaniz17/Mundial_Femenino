```python
import pandas as pd
```

```python
df_fixture = pd.read_csv('fifa_worldcup_fixture.csv')
df_data = pd.read_csv('fifa_worldcup_data.csv')
```

```python
df_fixture['home'] = df_fixture['home'].str.strip()
df_fixture['away'] = df_fixture['away'].str.strip()
```

```python
df_data[df_data['home'].isnull()]
```

```python
df_data['home']
```

```python
df_data = df_data.iloc[:579]
print(df_data)
```

```python
df_data.dropna(inplace=True)
```

```python
df_fixture.dropna(inplace=True)
```

```python
df_fixture
```

```python
print(df_data)
```

```python
df_data[df_data['score'].isnull()]
```

```python
df_data.drop_duplicates(inplace=True, ignore_index=True)
df_data
```

```python
# verify number of matches per competition
years = [1991,1995,1999,2003,2007,2011,2015,2019]
```

```python
for year in years:
    print(year, len(df_data[df_data['year']==year]))
```

```python
df_data['score'] = df_data['score'].str.replace('\([^()]*\)', '', regex=True)
df_data
```

```python
df_data[df_data['year']==2007]
```

```python
# splitting score columns into home and away goals and dropping score column
df_data[['HomeGoals', 'AwayGoals']] = df_data['score'].str.split(':', expand=True)
df_data.drop('score', axis=1, inplace=True)
```

```python
df_data
```

```python
df_data.rename(columns={'home': 'HomeTeam', 'away': 'AwayTeam', 
                                   'year':'Year'}, inplace=True)
df_data = df_data.astype({'HomeGoals': int, 'AwayGoals':int, 'Year': int})
```

```python
# creating new column "totalgoals"
df_data['TotalGoals'] = df_data['HomeGoals'] + df_data['AwayGoals']
df_data
```

```python
print(df_data[df_data['HomeTeam'].str.contains('Colombia')])
print(df_data[df_data['AwayTeam'].str.contains('Colombia')])
```

```python
df_data.to_csv('clean_fifa_worldcup_matches.csv',index=False)
df_fixture.to_csv('clean_fifa_worldcup_fixture.csv',index=False)
```
