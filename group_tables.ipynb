```python
all_tables = pd.read_html('https://es.wikipedia.org/wiki/Copa_Mundial_Femenina_de_F%C3%BAtbol_de_2023')

dict_table = {}
for letter, i in zip(alphabet, range(15,71,7)): # A=15, B=22, ..., H=64
    df = all_tables[i]
    df.rename(columns={df.columns[0]:'Team'}, inplace=True)
    dict_table[f'{letter}'] = df
```

```python
with open('dict_table', 'wb') as output:
    pickle.dump(dict_table, output)
```
